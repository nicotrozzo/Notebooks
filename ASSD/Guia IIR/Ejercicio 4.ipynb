4) Dadas las $H(s)$ se pide encontrar las $H(z)$ correspondientes utilizando el método invariante al impulso, y representar gráficamente el módulo de la respuesta en frecuencia de la $H(s)$ y la $H(z)$ en un mismo gráfico para comparar. Usar $\omega_s$ = $2\pi f_s$ = $40$. Para el gráfico usar dB para la ganancia y frecuencia lineal entre 0 y $f_s/2$
___

a) Se tiene:
\begin{equation}
    H_a(s) = \frac{8}{(s+2)(s+4)}
\end{equation}

La función verifica Gr(D) >= Gr(N) + 2, por lo que se puede utilizar el método invariante al impulso.

Descomponiendo en fracciones simples:
\begin{equation}
    H_a(s) = \frac{4}{s+2} - \frac{4}{s+4}
\end{equation}

Antitransformando Laplace, se obtiene la respuesta impulsiva:
\begin{equation}
    h_a(t) = 4(e^{-2t} - e^{-4t})u(t)
\end{equation}

Se la muestrea:
\begin{equation}
    h(nT) = 4(e^{-2nT} - e^{-4nT})u(n)
\end{equation}

Finalmente, se obtiene la transferencia digital transformando y multiplicando por T:
\begin{equation}
    H_D(z) = 4T \left( \frac{z}{z-e^{-2T}} - \frac{z}{z-e^{-4T}} \right) = \frac{4Tz(e^{-2T} - e^{-4T})}{(z-e^{-2T})(z-e^{-4T})}
\end{equation}

La respuesta en frecuencia digital es:
\begin{equation}
    H_D(e^{j\omega}) = 4T \left( \frac{e^{j\omega}}{e^{j\omega}-e^{-2T}} - \frac{e^{j\omega}}{e^{j\omega}-e^{-4T}} \right)
\end{equation}

```python
# Creamos las funciones transferencia analogica y digital

from scipy import signal
import numpy as np

# Transferencia analogica
zeros_a1 = []
poles_a1 = [-2, -4]
gain_a1 = 8
analog_tf1 = signal.ZerosPolesGain(zeros_a1, poles_a1, gain_a1)

# Transferencia digital
ws = 40
T = 2*np.pi/ws
zeros_d1 = [0]
poles_d1 = [np.exp(-2*T), np.exp(-4*T)]
gain_d1 = 4*(np.exp(-2*T) - np.exp(-4*T))*T
digital_tf1 = signal.ZerosPolesGain(zeros_d1, poles_d1, gain_d1, dt=T)
```

A continuación se muestra la comparación entre el módulo de la respuesta en frecuencia analógica y el de la digital, con la frecuencia desnormalizada:

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [6,4]
plt.rcParams.update({'font.size': 9})


w = np.linspace(0, ws/2, 1000)
w_a, mag_a, phase_a = analog_tf1.bode(w)
w_d, mag_d, phase_d = digital_tf1.bode()
plt.figure()
plt.plot(w_a, mag_a, label='Ha(jw)')
plt.plot(w_d, mag_d, label='Ha(exp(jw))')
plt.legend()
plt.title('Ha(jw) para el caso analogico y el digital')
plt.xlabel('Frecuencia (rad/s)')
plt.ylabel('Amplitud (dB)')
plt.show()
```

Se observa que el paso de analógico a digital con esta frecuencia de muestreo provoca aliasing, más perceptible al acercarse a $\omega_s/2$, donde afecta más el hecho de que $H_a(j\omega)$ no sea 0 más allá de $\omega_s/2$

b) Se tiene:
\begin{equation}
    H_b(s) = \frac{8}{s(s+2)(s+4)}
\end{equation}

Nuevamente se verifica Gr(D) >= Gr(N) + 2 y se puede utilizar el método invariante al impulso.

Descomponiendo en fracciones simples:
\begin{equation}
    H_b(s) = -\frac{2}{s+2} + \frac{1}{s+4} + \frac{1}{s}
\end{equation}

La respuesta impulsiva es:
\begin{equation}
    h_b(t) = (-2e^{-2t} + e^{-4t} + 1)u(t)
\end{equation}

Muestreando:
\begin{equation}
    h(nT) = (-2e^{-2nT} + e^{-4nT} + 1)u(n)
\end{equation}

Finalmente, se obtiene la transferencia digital transformando y multiplicando por T:
\begin{equation}
    H_D(z) = T \left( \frac{-2z}{z-e^{-2T}} + \frac{z}{z-e^{-4T}} + \frac{z}{z-1} \right) = 
     \frac{zT(1-e^{-2T})^2 \cdot (z+e^{-2T})}{(z-1)(z-e^{-2T})(z-e^{-4T})}
\end{equation}

```python
# Transferencia analogica
zeros_a2 = []
poles_a2 = [-2, -4, -0.00001] # se corre el polo del origen levemente para que no diverja
gain_a2 = 8
analog_tf2 = signal.ZerosPolesGain(zeros_a2, poles_a2, gain_a2)

# Transferencia digital
zeros_d2 = [0, -np.exp(-2*T)]
poles_d2 = [np.exp(-2*T), np.exp(-4*T), 0.9999] # se corre el polo levemente de |Z|=1 para que no diverja
gain_d2 = (1 - np.exp(-2*T))**2*T
digital_tf2 = signal.ZerosPolesGain(zeros_d2, poles_d2, gain_d2, dt=T)

w = np.linspace(0, ws/2, 1000) 
w_a, mag_a, phase_a = analog_tf2.bode(w)
w_d, mag_d, phase_d = digital_tf2.bode()
plt.figure()
plt.plot(w_a, mag_a, label='Hb(jw)')
plt.plot(w_d, mag_d, label='Hb(exp(jw))')
plt.legend()
plt.title('Hb(jw) para el caso analogico y el digital')
plt.xlabel('Frecuencia (rad/s)')
plt.ylabel('Amplitud (dB)')
plt.show()
```

En este caso se observa menos aliasing debido a que la caida del módulo de la respuesta en frecuencia es más abrupta por haber agregado el polo en el origen y ser la transferencia de mayor orden que $H_a$.

c) Se tiene:
\begin{equation}
    H_c(s) = \frac{s+1}{(s+0.5)(s+4)}
\end{equation}

En este caso no se verifica la condicion para utilizar el método invariante al impulso, se lo utilizará de todas formas
para observar qué pasa.
Recordamos que la condición de Gr(D) >= Gr(N) + 2 se pide para que $h_a(0^+) \approx 0$, ya que:
\begin{equation}
    H_D(e^{j\omega T}) = \frac{h_a(0^+)}{2} + 1/T \sum_{n=-\infty}^{+\infty} H_A(j\omega + jn\omega_s)
\end{equation}

Descomponiendo en fracciones simples:
\begin{equation}
    H_c(s) = -\frac{1/7}{s+0.5} + \frac{6/7}{s+4}
\end{equation}

La respuesta impulsiva es:
\begin{equation}
    h_c(t) = 1/7(e^{-0.5t} + 6e^{-4t})u(t)
\end{equation}

Notamos que:
\begin{equation}
    \frac{h_c(0^+)}{2} = 0.5
\end{equation}

Muestreando:
\begin{equation}
    h(nT) = 1/7(e^{-0.5nT} + e^{-4nT})u(n)
\end{equation}

Finalmente, se obtiene la transferencia digital transformando y multiplicando por T:
\begin{equation}
    H_D(z) = \frac{T}{7} \left( \frac{z}{z-e^{-0.5T}} + \frac{6z}{z-e^{-4T}} \right) = 
     \frac{zT \cdot (z-\frac{e^{-4T} + 6e^{-0.5T}}{7})}{(z-e^{-0.5T})(z-e^{-4T})}
\end{equation}

```python
# Transferencia analogica
zeros_a3 = [-1]
poles_a3 = [-0.5, -4]
gain_a3 = 1
analog_tf3 = signal.ZerosPolesGain(zeros_a3, poles_a3, gain_a3)

# Transferencia digital
zeros_d3 = [0, (np.exp(-4*T) + 6*np.exp(-0.5*T) )/7]
poles_d3 = [np.exp(-0.5*T), np.exp(-4*T)]
gain_d3 = T
digital_tf3 = signal.ZerosPolesGain(zeros_d3, poles_d3, gain_d3, dt=T)

w = np.linspace(0, ws/2, 1000) 
w_a, mag_a, phase_a = analog_tf3.bode(w)
w_d, mag_d, phase_d = digital_tf3.bode()
#w_d, H_d = digital_tf3.freqresp()
#mag_d = 20*np.log10(np.abs(H_d))
veces_d = np.power(10, mag_d/20)
mag_d_corregida = 20*np.log10(veces_d - 0.5*T)
plt.figure()
plt.plot(w_a, mag_a, label='Hc(jw)')
plt.plot(w_d, mag_d, label='Hc(exp(jw))')
plt.plot(w_d, mag_d_corregida, label='Hc(exp(jw)) - hc(0+)/2')
plt.legend()
plt.title('Hc(jw) para el caso analogico y el digital')
plt.xlabel('Frecuencia (rad/s)')
plt.ylabel('Amplitud (dB)')
plt.show()
```

Se muestra en este caso el módulo de la transferencia analógica, superpuesto a la transferencia digital. Como era de esperarse, como no se cumple que $h_c(0^+) \approx 0$, las transferencias analógica y digital difieren en una constante.

La tercer curva muestra el resultado de restarle a la transferencia digital el valor constante que se sabe que tiene sumado, y se puede ver que para bajas frecuencias se aproxima muy bien a la transferencia analógica. Sin embargo, al aumentar la frecuencia hay mucho aliasing, debido a que el valor de $H_c(s)$ no es despreciable más allá de $\omega_s/2$. 

Por último, cabe aclarar que el método invariante al impulso modificado también tendría un problema similar dado que sirve para grados de numerador y denominador >= 2, que en este caso no se cumple porque el grado del numerador es 1.
