```python
# Create PWL file
def create_pwl_file(a, b, filename):
    fileout = open(filename, "w")
    for k in range( np.amin([len(a), len(b)]) ):
        fileout.write(f'{a[k]:.8f} {b[k]:.8f}\n')
    fileout.close()
```

```python
# Señal cuadratica

import numpy as np
import matplotlib.pyplot as plt

# Periodo de la señal
f = 3500
T = 1/f

# Numero de periodos
n = 10  

# Espaciado entre valores
points_period = 1000
step = T/points_period

# Tension maxima
vmax = 5

x = np.arange(-T/2, (n-1/2)*T, step)


# y = vmax (x/(T/2))^2 entre -T/2 y T/2
y = [5*(2*t/T)**2 for t in x[:points_period]]


fig, ax1 = plt.subplots()
ax1.plot(x[:points_period], y)

ax1.set(xlabel='time (s)', ylabel='voltage (V)',
       title='One period')
ax1.grid()
plt.show()


y = np.tile(y,n)
fig, ax2 = plt.subplots()
ax2.plot(x[:3*points_period], y[:3*points_period])

ax2.set(xlabel='time (s)', ylabel='voltage (V)',
       title=f'{3} periods')
ax2.grid()
plt.show()

# Para la cuadratica
create_pwl_file(x[int(points_period/2):], y[int(points_period/2):], "Cuadratica3500Hz.txt")
```

```python
# Señal 3/2 seno

# Periodo de la señal
f = 3500
T = 1/f
# Numero de periodos
n = 10  

# Espaciado entre valores
points_period = 1000
step = T/points_period

# Tension maxima
vmax = 5

x = np.arange(0, n*T, step)

# Defino la salida en un periodo
y = [vmax/2*np.sin(3*np.pi*t/T) for t in x[:points_period] ] 

# Extension periodica
y = np.tile(y,n)

fig, ax1 = plt.subplots()
ax1.plot(x[:points_period], y[:points_period])

ax1.set(xlabel='time (s)', ylabel='voltage (V)',
       title='One period')
ax1.grid()
plt.show()

fig, ax2 = plt.subplots()
ax2.plot(x[:3*points_period], y[:3*points_period])

ax2.set(xlabel='time (s)', ylabel='voltage (V)',
       title=f'{3} periods')
ax2.grid()
plt.show()

# Para la cuadratica
create_pwl_file(x, y, "tresmediooseno3500Hz.txt")
```
