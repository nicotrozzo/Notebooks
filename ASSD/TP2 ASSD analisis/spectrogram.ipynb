```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

fo = 11025
fs = 44100
x = np.arange(0,4096)/fs
data = 10*np.sin(2*np.pi*fo*x)
f, t, Sxx = signal.spectrogram(data, fs, 'boxcar', 1000, 500, 1024)
plt.pcolormesh(t, f, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
# plt.plot(x, data)
plt.show()
```
