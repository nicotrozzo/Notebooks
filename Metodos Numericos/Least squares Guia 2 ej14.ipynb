```python
# Ej 14 Guia metodos Numericos
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, np.pi/4, np.pi])
y = np.array([0.97, 1.42, -1.04])
A = np.vstack([np.cos(x), np.sin(x)]).T
a, b = np.linalg.lstsq(A, y, rcond=None)[0]
print(f'a = {a:.5f}')
print(f'b = {b:.5f}')

t = np.linspace(-np.pi,np.pi,64)
plt.plot(x, y, 'o', label='Original data', markersize=10)
plt.plot(t, a*np.cos(t) + b*np.sin(t), 'r', label='Fitted curve')
plt.title(f'Least squares approximation, y={a:.3f}*cos(x)+{b:.3f}*sin(x)')
plt.legend()
plt.show()
```
